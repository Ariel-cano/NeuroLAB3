О задаче

Задача: Бинарная классификация.
Цель: На основе медицинских признаков пациента предсказать, есть ли у него диабет (Outcome = 1) или нет (Outcome = 0).
Целевая переменная:
Outcome (Диагноз диабета).
Признаки (Features):
Pregnancies (Беременности)
Glucose (Глюкоза)
BloodPressure (Артериальное давление)
SkinThickness (Толщина кожной складки)
Insulin (Инсулин)
BMI (Индекс массы тела)
DiabetesPedigreeFunction (Показатель наследственности по диабету)
Age (Возраст)

Выбор задачи бинарной классификации обоснован следующими ключевыми аспектами.
Тип целевой переменной является бинарным (значения 0 или 1), что прямо указывает на задачу классификации.
Содержательный контекст данных связан с медицинской диагностикой, где типичной является задача отнесения пациента к группе риска заболевания.
Структура данных представляет собой таблицу с признаками и меткой класса, что соответствует классическому формату для обучения с учителем.
При этом другие типы задач машинного обучения не применимы:
регрессия не подходит, поскольку целевая переменная не является непрерывной величиной;
кластеризация нецелесообразна, так как данные уже размечены (присутствует столбец Outcome), а цель состоит в прогнозировании, а не в поиске скрытых групп;
задача рекомендательных систем исключена ввиду отсутствия данных о пользователях, предметах или их взаимодействиях.
